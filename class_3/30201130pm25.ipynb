```python
import pandas as pd
dataset = pd.read_csv("30201130PM25.csv", header = None)
dataset.columns = ["pm25"]
```

```python
9999 in dataset.values
```

```python
import numpy as np
df_Imp = dataset.copy()
df_Imp['pm25'] = df_Imp['pm25'].replace([9999],np.nan)




```

```python
import pandas as pd
import plotly.graph_objects as go
from sklearn.impute import SimpleImputer

# Convert to numeric
df_Imp['pm25'] = pd.to_numeric(df_Imp['pm25'], errors='coerce')

# Create a Figure for Plotly
fig = go.Figure()

# Add trace for actual data with black color, solid style, and thicker line
fig.add_trace(go.Scatter(x=df_Imp.index, y=df_Imp['pm25'], mode='lines',
                         line=dict(color='black', dash='solid', width=8),
                         name='Actual Data'))

# Set x-axis range up to 500
fig.update_xaxes(range=[0, 500])

# Update layout
fig.update_layout(title='Plot-1', xaxis_title='Index', yaxis_title='pm25 Values')

# Show the plot
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from sklearn.impute import SimpleImputer

# Convert to numeric
df_Imp['pm25'] = pd.to_numeric(df_Imp['pm25'], errors='coerce')

# Perform mean imputation
imputer = SimpleImputer(strategy='mean')
sensorValues_mean = imputer.fit_transform(df_Imp['pm25'].values.reshape(-1, 1))

# Create a Figure for Plotly
fig = go.Figure()

# Add trace for mean imputation data with red color, solid style
fig.add_trace(go.Scatter(x=df_Imp.index, y=sensorValues_mean.flatten(), mode='lines',
                         line=dict(color='red', dash='solid'),
                         name='Mean Imputation'))

# Set x-axis range up to 500
fig.update_xaxes(range=[0, 500])

fig.update_layout(title='Plot-2 Graph', xaxis_title='Index', yaxis_title='pm25 Values')

# Show the plot
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from sklearn.impute import SimpleImputer

# Convert to numeric
df_Imp['pm25'] = pd.to_numeric(df_Imp['pm25'], errors='coerce')

# Perform mean imputation
imputer = SimpleImputer(strategy='mean')
sensorValues_mean = imputer.fit_transform(df_Imp['pm25'].values.reshape(-1, 1))

# Create a Figure for Plotly
fig = go.Figure()

# Add trace for actual data with red color and thicker line
fig.add_trace(go.Scatter(x=df_Imp.index, y=df_Imp['pm25'], mode='lines',
                         line=dict(color='black', width=8),
                         name='Actual Data'))

# Add trace for mean imputation with black color and normal line width
fig.add_trace(go.Scatter(x=df_Imp.index, y=sensorValues_mean.flatten(), mode='lines',
                         line=dict(color='red', width=2),
                         name='Mean Imputation'))

# Set x-axis range up to 500
fig.update_xaxes(range=[0, 500])

# Update layout
fig.update_layout(title='Original Data and Mean Imputation', xaxis_title='Index', yaxis_title='pm25 Values')

# Show the plot
fig.show()
```
